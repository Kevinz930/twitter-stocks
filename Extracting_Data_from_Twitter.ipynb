```python
#Import snscrape and pandas
import snscrape.modules.twitter as sntwitter
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', None)
```

```python
#Collect tweets from Twitter
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('tesla since:2020-01-01 until:2021-01-01').get_items()):
  if i>100:
    break
  tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.username])
   
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Writer'])
tweets_df
```
